## Seleccion de un conjunto de datos del dataset

```python
# Seleccionar una porcion de datos por posicion uso iloc.

titanic.iloc[[0,1],[2,3,4,5]]
```

```python
# Seleccionar datos por etiquetas: numero de row y nombre de columna uso loc

# Me traigo los datos de Name y SibSp de las filas 1 y 10 del dataset.

titanic.loc[[1,10],['Name', 'SibSp']]
```

```python
# Tipo Slices

#titanic.iloc[:0,:6] --> Considera la fila 0 como la fila del titulo.

# Para traerme 3 filas, incluyendo el titulo y 2 filas mas. Ademas, 7 columnas (de la 0 a la 6).

titanic.iloc[:2,:6]
```

```python
# Otro forma de traerme la data por Slices. Me traigo todas las columnas y las filas 0 a la 7 (la fila 0 es el titulo)

titanic.iloc[:6]
```

```python
# Traerme las ultimas 4 filas

titanic.iloc[-4:]
```

```python
# Traerme todas las filas menoslas ultimas 4

titanic.iloc[:-4]
```

```python
# Usando loc. Me traigo hasta la fila 4, y de la columna Name a la Parch

titanic.loc[1:4,'Name':'Parch']
```
