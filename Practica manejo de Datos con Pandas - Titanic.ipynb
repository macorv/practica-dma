### Tipos de Indexado

Podemos traernos un subset de datos utilizando los siguientes comandos en `python`

Con `iloc` seleccionamos por posicion de fila y columna
```python
    data.iloc[[1,2,3],[4,5]]
```

con `loc` seleccionamos por etiqueta de fila y columna
```python
    data.loc[['row1','row2'],['columna1','columna2','columna3']]
```

````python
#Creo una copia del dataset y reemplazo el index
temp = data.copy()

# Creo el string 'elemento_' con una iteracion por cada index
 temp.index=['elemento_'+ str(i) for i in temp.index]

temp
````
