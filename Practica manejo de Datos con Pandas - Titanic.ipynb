## Agrupaciones y Tablas de Contingencia

### Agrupaciones

Las agrupaciones sirven para hacer calculos sobre un subconjunto de datos. Generalmente tienen 3 partes:

1. Definir los grupos (sobre los que se van a hacer los calculos).
2. Aplicar calculos sobre esos grupos definidos anteriormente.
3. Combinar los resultados.

````python
# Agrupar datos

agrupado = titanic.groupby('Pclass')
agrupado
````

````python
# Por lo que vemos, solo se creo un objeto de agrupamiento, pero no se hizo ningun calculo.
# Apliquemos un calculo

# Calculamos el promedio para todas las variables numericas agrupadas por Pclass
agrupado.mean()
````

````python
# Calculamos el promedio para la variable 'Survived'
agrupado.Survived.mean()

# Conclusion: los que viajaban en primera clase tenian mas posibilidades de sobrevivir que los que estaban en 2da y 3ra
````

````python
# Calculo una serie de medidas de agregacion Media, Varianza, una funcion ad-hoc y Conteo.
agrupado.Survived.agg({'Promedio': 'mean', 'Varianza': 'var', 'Promedio2': np.mean, 'Cantidad': 'count'})
````

### Tablas de Contingencia

Las tablas de contingencia son una especie de tabla dinamica de Excel. Sirven para relacionar datos.

Se usa la funcion `crosstab`
```
pd.crosstab(column1, column2)
```

````python
pd.crosstab(titanic.Pclass, titanic.Survived)
````

````python
# En % usando la clase. 

pd.crosstab(titanic.Pclass, titanic.Survived).apply(lambda x: x/x.sum(), axis=1)

# Interpretacion:
# De los que viajaron en 1era, el 63% sobrevivio
# De los que viajaron en 2da, el 47% sobrevivio
# De los que viajaron en 3era, el 24% sobrevivio
````

````python
# En % usando si sobrevivio o no

pd.crosstab(titanic.Pclass, titanic.Survived).apply(lambda x: x/x.sum(), axis=0)

# Interpretacion:
# El 68% de los que no sobrevivieron pertenecen a tripulantes de la 3era clase
# El 40% de los que sobrevivieron pertenecer a la 1era clase.
````

````python
titanic.Age.value_counts()
````

````python
titanic.Age.value_counts(True).sort_index()
````

````python
pd.crosstab(titanic.Pclass, pd.cut(titanic.Age, [i * 10 for i in range(9)]), values=titanic.Survived, aggfunc=np.mean)
````

`pd.cut` crea intervalos


`[i * 10 for i in range(9)]` es lo mismo que:

```for i in range(9):
        i*10
```

Lo que hace es crear los intervalos para poner la edad en los cajones

** Interpretacion **

Los valores que se agrupan son los de Survived por edad. Por ejemplo:
1. Los chicos entre 0 y 10 anios de 2da clase se salvaron todos.

## Modelo

Separo los datos en entrenamiento y validacion

````python
# El 10% de validacion que habiamos armado antes
valid_index
````

````python
titanic.loc[valid_index]
````

````python
# Al dataset de entrenamiento tengo que sacarle la variable objetivo, es decir, 'Survived'

train = titanic.drop(valid_index, axis=0)

valid = titanic.loc[valid_index]
````

````python
X_train, y_train = train.drop('Survived', axis = 1), train.Survived
X_valid, y_valid = valid.drop('Survived', axis = 1), valid.Survived
````

````python
# Llamo a las librerias necesarias para armar el modelo

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import ParameterGrid

from sklearn.metrics import roc_auc_score, accuracy_score
````

````python
# Ayuda accuracy_score
accuracy_score?
````

````python
# Creo un array con diccionario adentro para cada modelo y los parametros que hay que llamar

learners = [
    
    {
        'learner': LogisticRegression,
        'params': {
            'C': [0.0001, 0.01, 1, 10, 100]
        }
    },
    {
        'learner': RandomForestClassifier,
        'params': {
            'max_depth': [4,10,20,None],
            'n_estimators': [10,100]
        }
    }
]
````

````python
pd.options.display.max_rows = 20

# Creo un array vacio
resultados = []

# Este es el for en el que prueba cada modelo y los parametros por cada modelo
for candidate in learners:
    for params in ParameterGrid(candidate["params"]):
        learner = candidate["learner"](**params)
        learner.fit(X_train, y_train)
        probs = learner.predict_proba(X_valid)[:, -1]
        if hasattr(learner, "feature_importances_"):
            print("importancia", params)
            print(pd.Series(learner.feature_importances_, index=X_train.columns))
            print("*"*10)
        elif hasattr(learner, "coef_"):
            print("coefs", params)
            print(pd.Series(learner.coef_[0], index=X_train.columns))
            print("*"*10)
        res = {
            "learner": learner.__class__.__name__,
            "parmas": params,
            "area": roc_auc_score(y_valid, probs),
            "accuracy": accuracy_score(y_valid, probs >= 0.5)
        }
        resultados.append(res)
````

````python
# Armo un dataframe con los resultados

resultados = pd.DataFrame.from_dict(resultados)[["learner", "parmas", "accuracy", "area"]]
````
